```python
"""
    This code block calculates the number of tag in a training set
"""
import pandas as pd
import os
import glob
import numpy as np
import multiprocessing
import string
import random
import time
import bs4

# use glob to get all the csv files 
# in the folder
csvDirectory = r"/home/muddi004/muddi/GIANT/downsized-4600/"
csv_files = glob.glob(os.path.join(csvDirectory, "*.csv"))
print(len(csv_files))

start = time.time()
print("Starting now")
print(len(csv_files))

sumofAuthor = 0
sumofCT = 0
sumofDate = 0
sumofPage = 0
sumofTitle = 0
sumofPublisher = 0
sumofVol = 0

for csv in csv_files:
    #print("Now working on: {}".format(csv))        
    dataset = pd.read_csv(csv)    
    for index, row in dataset.iterrows():
        citationString = row["citationStringAnnotated"].strip()
        soup = bs4.BeautifulSoup(citationString, "lxml")
        
        author = soup.find('author')
        ct = soup.find('container-title')
        vol = soup.find('volume')
        date = soup.find('issued')
        page = soup.find('page')
        title = soup.find('title')
        publisher = soup.find('publisher')
        issue = soup.find('issue')
        
        if author:
            sumofAuthor = sumofAuthor + 1
        if ct:
            sumofCT = sumofCT + 1
        if vol:
            sumofVol = sumofVol + 1
        if date:
            sumofDate = sumofDate + 1
        if page:
            sumofPage = sumofPage + 1
        if title:
            sumofTitle = sumofTitle + 1
        if publisher:
            sumofPublisher = sumofPublisher + 1        
    #print('Done Writing {}'.format(csv))

print("Total author: {}".format(sumofAuthor))
print("Total ct: {}".format(sumofCT))
print("Total date: {}".format(sumofDate))
print("Total page: {}".format(sumofPage))
print("Total title: {}".format(sumofTitle))
print("Total publisher: {}".format(sumofPublisher))
print("Total vol: {}".format(sumofVol))

end = time.time()
print("Total time required: {}".format(end - start))


```
